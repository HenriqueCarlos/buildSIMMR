```python
import xml.etree.ElementTree as ET
import os
from tabulate import tabulate
```

```python
#2.1 entity creation
commndIng = "create_ing("

#2.2 action
commndComb = "combine("
commndSep = "separate("

#2.3 location - no implementation
#put
#remove
#index of ingredients remain the same

#2.4 alter ingredients
commndCut = "cut("
commndMix = "mix("
commndCook = "cook("
commndDo = "do("

#Ingredient map

class Food:
    
    # count food created
    # also serve as index for the node
    foodCount = 0
    def __init__(self,fNm,fNo):
        self.fNm = self.normalizeFoodName(fNm)
        self.foodNo = fNo
        self.fIndex = Food.foodCount
        Food.foodCount += 1

    # set name and number
    # index with respect to the adjacency list however cannot be changed
    def setName(self, newFoodName):
        self.fNm = self.normalizeFoodName(newFoodName)
    
    def setFoodNo(self, newFoodNo):
        self.fNm = newFoodNo

        
    def getName(self):
        return self.fNm
    def getNo(self):
        return self.foodNo
    def getIndex(self):
        return self.fIndex
    
    def __str__(self):
        return self.fNm

#         return str(self.foodNo)+":"+self.fNm
        
    def normalizeFoodName(self,foodName):
        # To be implemented
        if (foodName[-1] != '\"'):
            foodName += '\"';
        
        return foodName

class IngMap:

    def __init__(self):
        self.m = {}
        self.adjList = []
        self.states = []
        self.sortedList = []
    
         
    # add Ingredients
    def addIng(self,ingNo, desc):
        thisFood = Food(desc,ingNo)
        self.m[ingNo] = thisFood
        self.adjList.append([])
    
    def modifyFood(self, ingNo, newIngNo, newIngDesc):
        oldFood = self.m[ingNo]
        self.m.pop(ingNo, None)
        oldFood.setFoodNo(newIngNo)
        oldFood.setName(newIngDesc)
        self.m[newIngNo] = oldFood
        
    
    # get Index
    def getIndex(self,ingNo):
        return self.m[ingNo].fIndex
    
    # link a child node with list of parent nodes
    def linkIng(self,ingChild, ingParents):
        indexForChild = self.getIndex(ingChild)
        for eachIngParent in ingParents:
            self.adjList[self.getIndex(eachIngParent)].append(indexForChild)

    
    def prepTopSort(self):
        self.states = [0 for each in range(len(self.adjList))]
        self.sortedList = []
        
    def topSort(self,nodeIndex):
        
        if (self.states[nodeIndex] != 0):
            return
        

        myChildren = self.adjList[nodeIndex]

        for each in myChildren:
            self.topSort(each)
        self.states[nodeIndex] = 1
        
        self.sortedList.append(nodeIndex)

    def getTopSort(self):
        self.prepTopSort()
        for eachIndex in range(len(self.states)):
            self.topSort(eachIndex)
        self.sortedList.reverse()        
        return self.sortedList


    
```

```python
def checkCommnd(text, commndKey):
    return text[:len(commndKey)] == commndKey

def getArgs(text):
    for charIndex in range(len(text)):
        if (text[charIndex] == '('):
            break
#     print "arg: ", text[(charIndex+1):-1]
    return text[(charIndex+1):-1]


def isIng(text):
    return checkCommnd(text,commndIng)

def interpretIng(text):
    # create_ing(ing28, "1/2 cup butter, softened")
    
    global ingredientMap

    ingArg = getArgs(text)
    ingAndDesc = ingArg.split(',')
    ingNo = getIngNum(ingAndDesc[0])
    ingDesc = ','.join(ingAndDesc[1:])
    ingredientMap.addIng(ingNo,ingDesc)     

def isComb(text):
    return checkCommnd(text,commndComb)

def interpretComb(text):    
    # combine(ingredient set ins, ingredient out, string outdesc, string manner)
    # combine({ing0, ing1}, ing2, "dough", "")
    
    global ingredientMap

    ingArg = getArgs(text)
    left = text.split('{')
    assert len(left) == 2, 'there are more than 1 {'
    right = left[1].split('}')
    assert len(right) == 2, 'there are more than 1 )'
    
    # get ingredient source
    setIngIn = right[0]
    setIngList = setIngIn.split(',')
    setIngListNo =  [getIngNum(eachIng) for eachIng in setIngList]
    
    remArgs = right[1].lstrip(',')
    remArgs = remArgs.split(',')
    
    # create new food
    ingOutNo = getIngNum(remArgs[0])
    ingOutDesc = remArgs[1]
#     print ingOutNo,ingOutDesc
    ingredientMap.addIng(ingOutNo,ingOutDesc)     

    # link the child food to new food created
    ingredientMap.linkIng(ingOutNo, setIngListNo)

def isSep(text):
    return checkCommnd(text,commndSep)

def interpretSep(text):
    # separate(ing0, ing1, "raisins", ing2, "peanuts and almonds", "")
    
    global ingredientMap
    
    ingArg = getArgs(text)
    ingArgs = ingArg.split(",")
    
    # ingredientNum of the source
    ing0No = getIngNum(ingArgs[0])
    
    # build new ingredients
    ing1No = getIngNum(ingArgs[1])
    ing1Des = ingArgs[2]
    ingredientMap.addIng(ing1No,ing1Des)  
    
    ing2No = getIngNum(ingArgs[3])
    ing2Des = ingArgs[4]
    ingredientMap.addIng(ing2No,ing2Des)  

    # from source ing0 separated (direct edges) to two others 1 and 2
    ingredientMap.linkIng(ing1No, [ing0No])
    ingredientMap.linkIng(ing2No, [ing0No])

def isCut(text):
    return checkCommnd(text,commndCut)

def interpretCut(text):
    # cut(ingredient in, tool t, ingredient out, string outdesc, string manner)
    global ingredientMap

    ingArg = getArgs(text)
    ingArgs = ingArg.split(",")

    ingNoExist = getIngNum(ingArgs[0])
    ingNoTransformed = getIngNum(ingArgs[2])
    desTransformed = ingArgs[3]
    
    ingredientMap.modifyFood(ingNoExist, ingNoTransformed, desTransformed)    

def isMix(text):
    return checkCommnd(text,commndMix)

def interpretMix(text):
    
    # mixing here does not add more ingredients (just a manner of cooking)
    # interchange old ingredient number with a new one
    # mix(ingredient in, tool t, ingredient out, string outdesc, string manner)
    
    global ingredientMap

    ingArg = getArgs(text)
    ingArgs = ingArg.split(",")
    
    ingNoExist = getIngNum(ingArgs[0])
    ingNoTransformed = getIngNum(ingArgs[2])
    desTransformed = ingArgs[3]
    
    ingredientMap.modifyFood(ingNoExist, ingNoTransformed, desTransformed)
    
def isCook(text):
    return checkCommnd(text,commndCook)

def interpretCook(text):
    # cook(ingredient in, tool t, ingredient out, string outdesc, string manner)
    # cook(ing0, t0, ing1, "hot water", "boil")
    
    global ingredientMap

    ingArg = getArgs(text)
    ingArgs = ingArg.split(",")
    
    ingNoExist = getIngNum(ingArgs[0])
    ingNoTransformed = getIngNum(ingArgs[2])
    desTransformed = ingArgs[3]
    
    ingredientMap.modifyFood(ingNoExist, ingNoTransformed, desTransformed)

def isDo(text):
    return checkCommnd(text,commndDo)

def interpretDo(text):
    
    # do(ing0, , ing1, "dumplings", "fold into dumpling shapes")
    
    global ingredientMap

    ingArg = getArgs(text)
    ingArgs = ingArg.split(",")
    
    ingNoExist = getIngNum(ingArgs[0])
    ingNoTransformed = getIngNum(ingArgs[2])
    desTransformed = ingArgs[3]
        
    ingredientMap.modifyFood(ingNoExist, ingNoTransformed, desTransformed)
    
    # debug
#     for each in ingredientMap.m:
#         print "key: ", each, ingredientMap.m[each].getIndex(), ingredientMap.m[each].getNo()
    


def getIngNum(text):
    text = text.lstrip()
    assert text[:3] == "ing"
    textSplit = text.split("ing")
    return int(textSplit[1])
```

```python
def buildGraph(recipeFile):
    
    verbose = False

    tree = ET.parse('/Users/Mick/Desktop/FoodCapstone/annotated_recipes/'+recipeFile)
    root = tree.getroot()

    # get data from xml file
    annotations = []
    originalTexts = []

    for child in root:
        
        # get annotations
        annotations.append(child.find('annotation').text)
        
        # get unique texts
        thisText = child.find('originaltext').text
        if not thisText in originalTexts:
            originalTexts.append(child.find('originaltext').text)
        
        if verbose:
            print child.find('annotation').text

    # parse the data

    for eachAn in annotations:
        if isIng(eachAn):
            if verbose:
                print "ing", eachAn
            interpretIng(eachAn)

        elif isComb(eachAn):
            if verbose:
                print "comb", eachAn
            interpretComb(eachAn)

        elif isMix(eachAn):
            if verbose:
                print "mix", eachAn
            interpretMix(eachAn)

        elif isCut(eachAn):
            if verbose:
                print "cut", eachAn
            interpretCut(eachAn)

        elif isDo(eachAn):
            if verbose:
                print "do", eachAn
            interpretDo(eachAn)

        elif isCut(eachAn):
            if verbose:
                print "cut", eachAn
            interpretCut(eachAn)

        elif isCook(eachAn):
            if verbose:
                print "cook", eachAn
            interpretCook(eachAn)
        
        elif isSep(eachAn):
            if verbose:
                print "separate", eachAn
            interpretSep(eachAn)
    


    # Display graph
    # list of foods
    foodNodes = ingredientMap.m.values()
    # table for display
    displayTable = [ [row,"nameFood",[]] for row in range(len(foodNodes)) ]
    for eachFood in foodNodes:
        displayTable[eachFood.getIndex()][1] = eachFood.getName()
    for i,each in enumerate(ingredientMap.adjList):
        stringChild = [str(eachChild) for eachChild in each ]
        stringChild = ",".join(stringChild)
        displayTable[i][2] = stringChild
    print tabulate(displayTable, headers=["node-id","node-form", "child-id"])
    
    return originalTexts, ingredientMap.adjList

    
path = '/Users/Mick/Desktop/FoodCapstone/annotated_recipes/'

countRep = 0
for filename in os.listdir(path):
    if filename[-4:] == ".xml":
        print "Recipe#: ", countRep
        recipeName = filename.split(".")
        print "Recipe Name: ", recipeName[0]
        print "SIMMR: "
        ingredientMap = IngMap()
        Food.foodCount = 0
        oriText, adjList = buildGraph(filename)
        
        print
        print "Original Texts: "
        for eachOriText in oriText:
            print eachOriText
        
#         for eachNode in adjList:
#             print eachNode
        
        print "\n\n########## end of recipe ", countRep, "##########\n"
        countRep += 1

```
