```python
def dbSpotLightAnnotate(text,conf,support,types):
#     print text
#     text = "package frozen mixed berries, sliced peaches, drained 2 tablespoons honey. In a blender, combine frozen fruit, canned fruit and honey. Blend until smooth. Add chicken and pork." 
    conf = 0.1
    support = 20
    types = "Freebase:/food/food,Freebase:/food/ingredient"
    parameters = {}
    parameters["text"] = text
    parameters["confidence"] = conf
    parameters["support"] = support
    parameters["types"] = types
    headers = {"Accept":'application/json'}
    r = requests.get('http://spotlight.dbpedia.org/rest/annotate', params = parameters, headers=headers)
    queriedRes = r.json()
#     print "got data"
#     print queriedRes
    concept_pos = []
    for eachTag in queriedRes[u'Resources']:
        originalWord = eachTag[u'@surfaceForm']
        dbConcept = eachTag[u'@URI'].split("/")[-1]
        pos = eachTag[u'@offset']
        concept_pos.append((originalWord,dbConcept,pos))
#         print originalWord, dbConcept,pos
#         print eachTag[u'@URI']
    return concept_pos
    
```

```python
from nltk.parse.stanford import StanfordParser
english_parser = StanfordParser('/Users/Mick/Documents/workspace/stanford-parser-full-2014-10-31/stanford-parser.jar', '/Users/Mick/Documents/workspace/stanford-parser-full-2014-10-31/stanford-parser-3.5.0-models.jar')
def stanfordTag(sentence):
    global english_parser
    tree = english_parser.raw_parse_sents(("",sentence))
    return tree[0]

def getNounsDes(POSTree):

    nouns = []
    for stree in POSTree.subtrees():
        if (stree.label() == "NP" and stree.height() == 3):
            pos = stree.pos()
            numTags = len(pos)
            j = 0
            while (j < numTags):
                newWord = []
                if (pos[j][1][:2] == "NN" or (pos[j][1][:2] == "JJ" and j != numTags-1)):
                    newWord.append(pos[j][0])
                    j += 1
                    if (j >= numTags):
                        nouns.append(" ".join(newWord))
                        break
                    while (pos[j][1][:2] == "NN"):
                        newWord.append(pos[j][0])
                        j += 1
                        if (j >= numTags):
                            break
                    nouns.append(" ".join(newWord))
                else:
                    j += 1
    return nouns
        
    
```
