```python
def printLimit(item, num):
    if (num == 0):
        for each in item:
            print (each)
        return
    
    count = 0
    for each in item:
        count += 1
        print (each)
        if count > num:
            break

def getDictFreq(listStuff):
    dictStuff = {}
    for each in listStuff:
        if( not each in dictStuff):
            dictStuff[each] = 1
        else:
            dictStuff[each] += 1
    
    sortedStuff = sorted(dictStuff.items(), key=operator.itemgetter(1))
    
    return dictStuff, sortedStuff

def removeThesePOS(sentence, POStoRemoved):

    tokens = word_tokenize(sentence)
    tagged = pos_tag(tokens)
#     print tagged
    remainTok = [d[0] for d in tagged if not d[1] in POStoRemoved]
    return " ".join(remainTok)

class PhraseMaker:
    def __init__(self, foodDict, wordDict, minFreq, myStem):
            foodTok = foodDict.keys()
            wordTok = wordDict.keys()
            wordTok = [tk for tk in wordTok if wordDict[tk] >= minFreq]
            allTok = foodTok + wordTok
            tokPosition = {}
            for i, tk in enumerate(allTok):
                tokPosition[tk] = i
                
            self.tokPosition = tokPosition
            self.foodTok = foodTok
            self.wordTok = wordTok
            self.myStem = myStem
            self.sizeRep = len(allTok)
            self.keyVec = self.getKeyVec()
            self.sizeVec = len(allTok)+2
    
    def makePhrase(self, phrase):
        
        inFood = 0
        inWord = 0
        vec = [0 for i in range(self.sizeRep)]

        
        if len(phrase) == 0:
            vec += [inFood, inWord]
            return vec
        
        
        phrase = phrase.split()
        phraseStemmed = [myStem.stem(each) for each in phrase]
        
        for each in phraseStemmed:
            if each in self.tokPosition:
                if each in self.foodTok:
                    inFood = 1
                if each in self.wordTok:
                    inWord = 1
                    
                vec[self.tokPosition[each]] = 1
                
        vec += [inFood, inWord]
        
#         print phrase, "- food: ", inFood, " word: ", inWord        
        return vec

    def getKeyVec(self):
        
        vec = ["" for i in range(self.sizeRep)]
        for each in self.tokPosition:
            vec[self.tokPosition[each]] = each
        
        vec += ["inFood", "inWord"]
        
        return vec
    
    def showKeySparse(self, vec):
        
        assert len(self.keyVec) == len(vec)
        for i, val in enumerate(vec):
            if (val != 0):
                print self.keyVec[i],":",val
    
```

```python
## build word bank without ing
#Main
countRep = 0
pointCheck = 0
toCheck = []
allWords = []
firstWords = []
verbose = 0

allFiles = os.listdir(path)
allFiles = [f for f in allFiles if f[-4:] == ".xml"]
allFiles = [allFiles[ri] for ri in randomOrder]
for filename in allFiles:        
        if toCheck:
            if pointCheck >= len(toCheck):
                break
            if (toCheck[pointCheck] != countRep):
                countRep += 1
                continue
            else:
                pointCheck += 1

        x, y = mngTgDat.getXandY(filename)
        ingString, xRem, yRem = mngTgDat.separateIngInstruct(x,y)

        xInstructGen, yInstructGen = mngTgDat.getOneToOneMatch(xRem,yRem, [mngTgDat.isGeneral])
        
        lenInstruct = len(xInstructGen)
        for i, each in enumerate(xInstructGen):
            ## find top words
            eachInstruct = each.rstrip().rstrip(".")
            instructTok = word_tokenize(eachInstruct)
            instructTok = [myStem.stem(tk) for tk in instructTok]
    
            withOutFirstWord = instructTok[1:]
            nonFoodTok = [etk for etk in withOutFirstWord if each not in ingDict]
            allWords += nonFoodTok
            
            ## get directions
            firstWords.append(instructTok[0])
                    
        if not toCheck:
            if (countRep == trainSize):
                break
        countRep += 1
        
if toCheck:
    trainSize = len(toCheck)
    
# allWords = [myStem.stem(each) for each in allWords]
wordDict, sortedStuff = getDictFreq(allWords)
firstWordDict, firstWordSort = getDictFreq(firstWords)
```

```python
minFreq = 8
phrsMker = PhraseMaker(ingDict, wordDict, minFreq, myStem)
minFreqFirstWord = 3
phrsMkerFirstWord = PhraseMaker({},firstWordDict,minFreqFirstWord,myStem)
```

```python
#build vector
countRep = 0
pointCheck = 0
toCheck = []

allX =[]
allY =[]
for filename in allFiles:        
 
        if toCheck:
            if pointCheck >= len(toCheck):
                break
            if (toCheck[pointCheck] != countRep):
                countRep += 1
                continue
            else:
                pointCheck += 1
        
        x, y = mngTgDat.getXandY(filename)
        xIng, xRem, yRem = mngTgDat.separateIngInstruct(x,y)
        xInstruct, yInstruct = mngTgDat.getOneToOneMatch(xRem,yRem, [mngTgDat.isGeneral])
        
        ## Labeled
        numberedInstrctXWithOri, ingMap, links = mngTgDat.processLabeledData(filename)
        
        ## pairing instruct
        instructList = ingMap.instruct
        
        if verbose == 0:
            print "Recipe#: ", countRep
            recipeName = filename.split(".")
            print "Recipe Name: ", recipeName[0]
            print
            for i, eIng in enumerate(xIng):
                print "ING#"+str(i), eIng

            print
        
        for l in links:
            print l

        ingList = []
        for eachIngProc in xIng:
            eachIngProc = prsIngHp.removeUnitsNum(myStem, eachIngProc, units)
            eachIngProc = removeThesePOS(eachIngProc, ['TO','IN','CONJ', 'CC',])
            ingList.append(eachIngProc)
        
        ## normalizd freq of word in instructions
        localWords = []
        totalWords = 0
        for eStruct in xInstruct:
            eStruct = eStruct.rstrip().rstrip(".")
            eStructSplit = word_tokenize(eStruct)
            totalWords += len(eStructSplit)
            localWords += [myStem.stem(eTk) for eTk in eStructSplit]
        localDict, localSorted = getDictFreq(localWords)
        
        
        for eKey in localDict:
            localDict[eKey] /= float(totalWords)
        
        Xdat = []
        Ydat = []
        
        ## interaction with Ing
        for i, eachIngProc in enumerate(ingList):
            if verbose == 1:
                print "ING#"+str(i), eachIngProc,"\n"
            
            relPosIng = i/float(len(ingList))
            interactIngFeats = []
            sentenceFeats = []
            labels = []
            
            ## per sentence
            for j, eachInstruct in enumerate(xInstruct):
                
                ## get label
                label = 0
                if j in numberedInstrctXWithOri:
                    allIngs = numberedInstrctXWithOri[j]
                    if i in allIngs[2]:
                        if verbose == 1:
                            print "********CORRECT********"
                            print  j+len(ingList) ,eachInstruct , "should be in ", instructList[i]

                        label = 1
                        
                        assert j+len(ingList) in instructList[i]
                            
                        
                labels.append(label)
                    
                ## get feature
                getFeat = 1
                if getFeat == 1:
                
                    relPosSen = j/(float(len(xInstruct)))
                    finalRes = prsIngHp.checkNGramMatch(myStem, eachInstruct,[eachIngProc])
                    ingSet = prsIngHp.getOriginWordFromSentence(eachInstruct, [(each[1],each[2],each[3],each[4]) for each in finalRes])
                    beforeWord, afterWord = prsIngHp.getBeforeAfterWord(eachInstruct, [(each[1],each[2],each[3],each[4]) for each in finalRes])

                    ingFoundAll = []
                    beforeAll = []
                    afterAll = []

                    freqScore = 0
                    rawScore = 0
                    freqAdjScore = 0
                    numHitUnique = 0
                    maxNgram = 0
                    similarPos = 0
                    corFreqEarly = 0

                    ## Sentence level feature


                    ## get feature of the interaction between ingredient and the sentence
                    for k, eachIngFound in enumerate(ingSet):

                        ## 1) Weighted Relative Frequency of the phrase found
                        eIngSplit = eachIngFound.split()
                        freqScore = sum([localDict[myStem.stem(eTk)] for eTk in eIngSplit]) / float(len(eIngSplit))

                        ## 2) Frequency adjusted max score
                        ## 3) Max raw score
                        tempScore = fuzz.ratio(eachIngFound,ingList[i])
                        invrsFreq = 1.0/freqScore
                        tempAdjScore = tempScore*invrsFreq
                        if (freqAdjScore < tempAdjScore):
                            rawScore = tempScore
                            freqAdjScore = tempAdjScore

                        ## get the word, and contexts
                            ingFoundAll = [eachIngFound]
                            beforeAll = [beforeWord[k]]
                            afterAll = [afterWord[k]]

    #                         print eachIngFound, "VS" , ingList[i], "-", tempScore
    #                         print "freqScore: ", freqAdjScore

                    ## 4) Number of hits from the sentence and ingredients
                    numHitUnique = len(set([myStem.stem(eIng) for eIng in ingSet]))

                    ## 5) Max N-gram for this sentence and ingredient
                    if (ingSet):
                        maxNgram = max([len(eIng.split()) for eIng in ingSet])

                    ## 6) How correlated is the position of ingredient and its appearance in the sentence
                    similarPos = 1- abs(relPosIng-relPosSen)

                    ## 7) How correlated is frequency of ingredient and its early appearance
                    corFreqEarly = 1 - abs(freqScore-relPosSen)

                    ## 8) Make the ingredient
                    ingMatched = phrsMker.makePhrase(" ".join(ingFoundAll))
                    # give score to isFood
                    ingMatched[-2] = ingMatched[-2]*freqAdjScore

                    ## 9) Make the context
                    contextMatched = phrsMker.makePhrase(" ".join(beforeAll + afterAll))

    #                 if (len(ingSet) != 0):
                    if verbose == 1:
                        print str(j)+')', eachInstruct, ingSet
                        print "words matched:", " ".join(ingFoundAll), "(", sum(ingMatched) ,")"
                        print "contexts:", " ".join(beforeAll + afterAll), "(",sum(contextMatched),")"
                        print "raw Score: ", rawScore
                        print "scoreFreqAdjust:", freqAdjScore
                        print "numHitUnique: ", numHitUnique
                        print "maxNgram: ", maxNgram
                        print "similar position:", similarPos
                        print "correl Freq_Early: ", corFreqEarly
                        print "foodValue: ",ingMatched[-2]
                        print

    #                 phrsMker.showKeySparse(ingMatched)
    #                 phrsMker.showKeySparse(contextMatched)

                    interactFeat = [
                        ## 1) Weighted Relative Frequency of the phrase found
                        freqScore, 
                        ## 2) Frequency adjusted max score
                        rawScore, 
                        ## 3) Max raw score
                        freqAdjScore, 

                        ## 4) Number of hits from the sentence and ingredients
                        numHitUnique, 

                        ## 5) Max N-gram for this sentence and ingredient
                        maxNgram, 

                        ## 6) How correlated is the position of ingredient and its appearance in the sentence
                        similarPos,

                        ## 7) How correlated is frequency of ingredient and its early appearance
                        corFreqEarly]

                        ## 8) Make the ingredient
                        ## 9) Make the context
                    interactFeat += ingMatched

                    interactIngFeats.append(interactFeat)

                    ## Sentence level feature

                    sentenceFeat = [] 
                    ## 10) relative position of the sentence
                    lenInstruct = len(xInstruct)
                    thisSentencePos = i/float(lenInstruct)

                    ## 11) Make first word: get direction
                    eachInstruct = eachInstruct.rstrip().rstrip(".")
                    instructTok = word_tokenize(eachInstruct)
                    instructTok = [myStem.stem(tk) for tk in instructTok]
                    direction = phrsMkerFirstWord.makePhrase(instructTok[0])

                    ## 12) Make the rest of the sentence
                    restOfSen = phrsMker.makePhrase(" ".join(instructTok[1:]))

                    ## 10) relative position of the sentence
                    sentenceFeat.append(thisSentencePos)
                    ## 11), 12)
                    sentenceFeat += direction + restOfSen

                    sentenceFeats.append(sentenceFeat)
                
            Ydat.append(labels)
            if verbose == 1:
                print "lenXdat", len(Xdat), len(interactIngFeats) , len(sentenceFeats)
            Xdat.append([interactIngFeats[ifeat]+ sentenceFeats[ifeat] for ifeat in range(len(interactIngFeats))])
        
        allX.append(Xdat)
        allY.append(Ydat)
        
        if verbose == 1:
            print "\n\n########## end of recipe ", countRep, "##########\n"
            
        if not toCheck:
            if (countRep == trainSize):
                break
        countRep += 1

        
if toCheck:
    trainSize = len(toCheck)

```
