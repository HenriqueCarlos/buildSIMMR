```python
import nltk
import operator
from string import punctuation
```

```python
def printLimit(item, num):
    if (num == 0):
        for each in item:
            print (each)
        return
    
    count = 0
    for each in item:
        count += 1
        print (each)
        if count > num:
            break

def isAllUpper(word):
    word = ''.join([ch for ch in word if ch not in punctuation])
    word = word.replace(" ","")
    return sum([ (eachLetter.isupper() or eachLetter.isdigit()) for eachLetter in word]) == len(word)

def getDictFreq(listStuff):
    dictStuff = {}
    for each in listStuff:
        if( not each in dictStuff):
            dictStuff[each] = 1
        else:
            dictStuff[each] += 1
    
    sortedStuff = sorted(dictStuff.items(), key=operator.itemgetter(1))
    
    return dictStuff, sortedStuff
     
def containTheseLetters(word, letters):
    for ch in word:
        if ch in letters:
            return True
    return False
    
```

```python
from nltk import stem
myStem = stem.snowball.EnglishStemmer()
ingF = open('/Users/Mick/Desktop/FoodCapstone/scirep-cuisines-detail/all_Ing.txt')
ingDict = {}
for line in ingF:
    ingFreq = line.split(",")
    thisIng = ingFreq[0]
    thisIng = [myStem.stem(each) for each in thisIng.split('_')]
#     for each in thisIng:
    each = thisIng[-1]
    if (each in ingDict):
        ingDict[each] += int(ingFreq[1].rstrip())
    else:
        ingDict[each] = int(ingFreq[1].rstrip())
```

```python
numPrint = 20
punctuationWOHi = punctuation.replace("-","")
fdFile = open('/Users/Mick/Desktop/FoodCapstone/USDAFood/sr27asc/FOOD_DES.txt')
foodDict = {}
foodWords = {}
foodList = []
validFood = {}
idsFood = []

USDANotYY = []
YYnotUSDA = []
for line in fdFile:
    thisID = int(line.split('~')[1])
    splitLine = line.split('~^~')
    foodDes = splitLine[2].split(",")
    thisFood = foodDes[0] # the actual food
    
    ## tokenize the food descriptions
    for i in range(1,len(foodDes)):
        if( not foodDes[i] in foodWords):
            foodWords[foodDes[i]] = 1
        else:
            foodWords[foodDes[i]] += 1
    
    ## dict of the food 
    if (not thisFood in foodDict):
        foodDict[thisFood] = 1
    else:
        foodDict[thisFood] += 1
    
    ## check with valid food from YY
    foodToken = [myStem.stem(each) for each in thisFood.split()]
    eachFTok = foodToken[0] # take the first word, most general item
    
    if eachFTok in ingDict:
        if not eachFTok in validFood:
            validFood[eachFTok] = [(",".join(foodDes[1:])).rstrip().lstrip()]
        else:
            validFood[eachFTok].append((",".join(foodDes[1:])).rstrip().lstrip())

        idsFood.append(thisID)
        
    else:
        firstWord = thisFood.split()[0].rstrip().lstrip()
        if (not isAllUpper(firstWord)): # all capital brand names
            if ( not containTheseLetters(firstWord,punctuationWOHi )): # has ' suggests brand
                if ("-" in firstWord): # check things in hiphen
                    thisFoodNoHiphen = firstWord.replace("-"," ")
                    text = nltk.word_tokenize(thisFoodNoHiphen)
                    tagged = nltk.pos_tag(text)
                    # check if any is verb combined
                    noTake = 0
                    for eachTag in tagged:
                        if (eachTag[1][:2] != "NN"):
                            noTake = 1
                    
                    # check if it ends with flavor
                    if (firstWord.split("-")[-1] == "flavor"):
                        noTake = 1
                    if (noTake == 0):
                        USDANotYY.append(firstWord)
                        idsFood.append(thisID)

                else:
                    text = nltk.word_tokenize(firstWord)
                    text = [each.lower() for each in text]
                    tagged = nltk.pos_tag(text)
                    noTake = 0
                    for eachTag in tagged:
                        if (eachTag[1][:2] != "NN"):
                            noTake = 1
                    if (noTake == 0):
                        USDANotYY.append(firstWord)
                        idsFood.append(thisID)

        

```

```python
unitFile = open('/Users/Mick/Desktop/FoodCapstone/USDAFood/sr27asc/WEIGHT.txt')
units = []
idsFood = list(set(idsFood))
for line in unitFile:
    if (line.split('~^')[1][0] != "1"):
        continue
    idsFoodHere = int(line.split('~')[1])
    
    if idsFoodHere in idsFood:
        splitLine = line.split('^~')
        splitLine = splitLine[1].split('~^')
        thisUnit = splitLine[0].replace(",",'').split()
        units.append(thisUnit[0])

dictUnit, sortUnit = getDictFreq(units)

USDANotYYSet = [myStem.stem(each) for each in USDANotYYSet]

filterdUnits = []
for each in sortUnit:
    try:
        thisStem = myStem.stem(each[0])
        if ( (thisStem in USDANotYYSet) or (thisStem in ingDict)):
            continue
        else:
            if (each[1] >= 4) and (not isAllUpper(each[0])):
                filterdUnits.append(each)
    except: 
        pass

for each in filterdUnits:
    print (each)
        
```
