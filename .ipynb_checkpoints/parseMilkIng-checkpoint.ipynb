```python
#Ingredient map
class Food:
    
    # count food created
    # also serve as index for the node
    foodCount = 0
    ingCount = 0
    def __init__(self,fNm,fNo):
        self.fNm = self.normalizeFoodName(fNm)
        self.foodNo = fNo
        self.fIndex = Food.foodCount
        Food.foodCount += 1
        self.composite = []
        self.raw = True

    # set name and number
    # index with respect to the adjacency list however cannot be changed
    def setName(self, newFoodName):
        self.fNm = self.normalizeFoodName(newFoodName)
    
    def setFoodNo(self, newFoodNo):
        self.fNm = newFoodNo
    
    def setProcess(self):
        self.raw = False

    def getName(self):
        return self.fNm
    def getNo(self):
        return self.foodNo
    def getIndex(self):
        return self.fIndex
    
    def __str__(self):
        return self.fNm

#         return str(self.foodNo)+":"+self.fNm
        
    def normalizeFoodName(self,foodName):
        # To be implemented
        foodName = foodName.lstrip()
        if (foodName[-1] != '\"'):
            foodName += '\"';
        if (foodName[0] != '\"'):
            foodName = '\"'+foodName
        
        return foodName
    
    def printComposite(self, comp, ingMap, depth):
        
        if len(comp) != 0:
            if (type(comp[0]) == type([])):
                for d in comp:
                    self.printComposite(d, ingMap, depth+1)
                print ""
            else:
                print comp[0],
        
class IngMap:

    def __init__(self):
        self.m = {} # key - ID according to milk label, value - Food object
        self.adjList = []
        self.states = []
        self.sortedList = []
        self.originIng = {}
        Food.foodCount = 0
        Food.ingCount = 0
    
    # add Ingredients
    def addIng(self,ingNo, desc):
        thisFood = Food(desc,ingNo)
        self.m[ingNo] = thisFood
        self.adjList.append([])
#         print "addIng:", self.m.keys()
#         print "originIng: ", [ self.originIng[e].getIndex() for e in self.originIng.keys()]
    
    def modifyFood(self, ingNo, newIngNo, newIngDesc):
        oldFood = self.m[ingNo]
        self.m.pop(ingNo, None)
        oldFood.setFoodNo(newIngNo)
        oldFood.setName(newIngDesc)
        self.m[newIngNo] = oldFood
        self.m[newIngNo].setProcess()
        
    # get Index
    def getIndex(self,ingNo):
        return self.m[ingNo].fIndex
    
    # link a child node with list of parent nodes
    def linkIng(self,ingChild, ingParents):
        indexForChild = self.getIndex(ingChild)
        for eachIngParent in ingParents:
            self.adjList[self.getIndex(eachIngParent)].append(indexForChild)

    def prepTopSort(self):
        self.states = [0 for each in range(len(self.adjList))]
        self.sortedList = []
        
    def topSort(self,nodeIndex):
        
        if (self.states[nodeIndex] != 0):
            return
        
        myChildren = self.adjList[nodeIndex]

        for each in myChildren:
            self.topSort(each)
        self.states[nodeIndex] = 1
        
        self.sortedList.append(nodeIndex)

    def getTopSort(self):
        self.prepTopSort()
        for eachIndex in range(len(self.states)):
            self.topSort(eachIndex)
        self.sortedList.reverse()        
        return self.sortedList

    # how many new nodes are constructed from this
    def getNumAddiNodes(self):
        return len(self.adjList) - len(self.originIng.keys())

    ## INTERPRET ##
    def interpretIng(self,text):
        # create_ing(ing28, "1/2 cup butter, softened")

        ingNo, ingString = mngTgDat.preProcessIngArgs(text)
        ingDesc, status = normalizeIng(ingString)
        self.addIng(ingNo,ingDesc)

        self.originIng[self.getIndex(ingNo)] = self.m[ingNo]
        self.m[ingNo].composite.append(self.getIndex(ingNo))

    def interpretComb(self,text):    
        # combine(ingredient set ins, ingredient out, string outdesc, string manner)
        # combine({ing0, ing1}, ing2, "dough", "")

        ingArg = mngTgDat.getArgs(text)
        left = text.split('{')
        assert len(left) == 2, 'there are more than 1 {'
        right = left[1].split('}')
        assert len(right) == 2, 'there are more than 1 )'

        # get ingredient source
        setIngIn = right[0]
        setIngList = setIngIn.split(',')
        setIngListNo =  [mngTgDat.getIngNum(eachIng) for eachIng in setIngList]

        remArgs = right[1].lstrip(',')
        remArgs = remArgs.split(',')

        # create new food
        ingOutNo = mngTgDat.getIngNum(remArgs[0])
        ingOutDesc = remArgs[1]
    #     print ingOutNo,ingOutDesc
        self.addIng(ingOutNo,ingOutDesc)     
#         print text
        oriIndex = [self.m[eachIng].getIndex() for eachIng in setIngListNo if ((len(self.m[eachIng].composite) == 1) and (self.m[eachIng].raw))]
#         print [self.m[eachIng].composite for eachIng in setIngListNo] 

        # link the child food to new food created
        self.linkIng(ingOutNo, setIngListNo)

        for ing in setIngListNo:
            self.m[ingOutNo].composite.append(self.m[ing].composite)
        
        if (oriIndex):
            return (text, oriIndex)
        
    def interpretSep(self,text):
        # separate(ing0, ing1, "raisins", ing2, "peanuts and almonds", "")

        ingArg = mngTgDat.getArgs(text)
        ingArgs = ingArg.split(",")

        # ingredientNum of the source
        ing0No = mngTgDat.getIngNum(ingArgs[0])

        # build new ingredients
        ing1No = mngTgDat.getIngNum(ingArgs[1])
        ing1Des = ingArgs[2]
        self.addIng(ing1No,ing1Des)  

        ing2No = mngTgDat.getIngNum(ingArgs[3])
        ing2Des = ingArgs[4]
        self.addIng(ing2No,ing2Des)  

        # from source ing0 separated (direct edges) to two others 1 and 2
        self.linkIng(ing1No, [ing0No])
        self.linkIng(ing2No, [ing0No])

        if self.m[ing0No].getIndex() in self.originIng:
            return (text, [self.m[ing0No].getIndex()] )



    def interpretCut(self,text):
        # cut(ingredient in, tool t, ingredient out, string outdesc, string manner)

        ingArg = mngTgDat.getArgs(text)
        ingArgs = ingArg.split(",")

        ingNoExist = mngTgDat.getIngNum(ingArgs[0])
        ingNoTransformed = mngTgDat.getIngNum(ingArgs[2])
        desTransformed = ingArgs[3]
        
        retVal = []
        if ((self.m[ingNoExist].getIndex() in self.originIng) and (self.m[ingNoExist].raw)):
            retVal = (text, [self.m[ingNoExist].getIndex()] )
        self.modifyFood(ingNoExist, ingNoTransformed, desTransformed)    
        if (retVal):
            return retVal
   
    def interpretMix(self,text):

        # mixing here does not add more ingredients (just a manner of cooking)
        # interchange old ingredient number with a new one
        # mix(ingredient in, tool t, ingredient out, string outdesc, string manner)


        ingArg = mngTgDat.getArgs(text)
        ingArgs = ingArg.split(",")

        ingNoExist = mngTgDat.getIngNum(ingArgs[0])
        ingNoTransformed = mngTgDat.getIngNum(ingArgs[2])
        desTransformed = ingArgs[3]

        retVal = []      
        if ((self.m[ingNoExist].getIndex() in self.originIng) and (self.m[ingNoExist].raw)):
            retVal = (text, [self.m[ingNoExist].getIndex()] )
        self.modifyFood(ingNoExist, ingNoTransformed, desTransformed)    
        if (retVal):
            return retVal
   
        
    def interpretCook(self,text):
        # cook(ingredient in, tool t, ingredient out, string outdesc, string manner)
        # cook(ing0, t0, ing1, "hot water", "boil")


        ingArg = mngTgDat.getArgs(text)
        ingArgs = ingArg.split(",")

        ingNoExist = mngTgDat.getIngNum(ingArgs[0])
        ingNoTransformed = mngTgDat.getIngNum(ingArgs[2])
        desTransformed = ingArgs[3]

        retVal = []      
        if ((self.m[ingNoExist].getIndex() in self.originIng) and (self.m[ingNoExist].raw)):
            retVal = (text, [self.m[ingNoExist].getIndex()] )
        self.modifyFood(ingNoExist, ingNoTransformed, desTransformed)    
        if (retVal):
            return retVal
        
    def interpretDo(self,text):

        # do(ing0, , ing1, "dumplings", "fold into dumpling shapes")

        ingArg = mngTgDat.getArgs(text)
        ingArgs = ingArg.split(",")

        ingNoExist = mngTgDat.getIngNum(ingArgs[0])
        ingNoTransformed = mngTgDat.getIngNum(ingArgs[2])
        desTransformed = ingArgs[3]

        retVal = []      
        if ((self.m[ingNoExist].getIndex() in self.originIng) and (self.m[ingNoExist].raw)):
            retVal = (text, [self.m[ingNoExist].getIndex()] )
        self.modifyFood(ingNoExist, ingNoTransformed, desTransformed)    
        if (retVal):
            return retVal
    
    def interpretPut(self,text):
        # put({ing9}, t0)
        if ("{" in text):
            ingArg = mngTgDat.getArgs(text)
            left = text.split('{')
            assert len(left) == 2, 'there are more than 1 {'
            right = left[1].split('}')
            assert len(right) == 2, 'there are more than 1 )'

            # get ingredient list that are put
            setIngIn = right[0]
            setIngList = setIngIn.split(',')
            setIngListNo =  [mngTgDat.getIngNum(eachIng) for eachIng in setIngList]

            retValIng = []      
            for ingNoExist in setIngListNo:
                if ((self.m[ingNoExist].getIndex() in self.originIng) and (self.m[ingNoExist].raw)):
                    retValIng.append(self.m[ingNoExist].getIndex())

            for ingNoExist in setIngListNo:
                self.m[ingNoExist].setProcess()

            if (retValIng):
                return (text,retValIng)
        else:
            ingArg = mngTgDat.getArgs(text)
            ingArgs = ingArg.split(",")

            ingNoExist = mngTgDat.getIngNum(ingArgs[0])

            retVal = []      
            if ((self.m[ingNoExist].getIndex() in self.originIng) and (self.m[ingNoExist].raw)):
                retVal = (text, [self.m[ingNoExist].getIndex()] )
            self.m[ingNoExist].setProcess()    
            if (retVal):
                return retVal
        
   
```

```python
def buildGraph(ingredientMap,recipeFile):
    
    instrctXWithOri = []
    verbose = False

    root = mngTgDat.getRoot('/Users/Mick/Desktop/FoodCapstone/annotated_recipes/',recipeFile)

    # get data from xml file
    annotations = []
    originalTexts = []
    allTexts = []
    dictTextEdge = {}

    for child in root:
        
        # get annotations
        annotations.append(child.find('annotation').text)
        
        # get unique texts
        thisText = child.find('originaltext').text
        if not thisText in originalTexts:
            originalTexts.append(child.find('originaltext').text)
        allTexts.append(thisText)
        
        if verbose:
            print child.find('annotation').text

    # parse the data

    for i, eachAn in enumerate(annotations):
        eachAn = eachAn.encode('ascii', 'ignore')
        if mngTgDat.isIng(eachAn):
            if verbose:
                print "ing", eachAn
            retVal = ingredientMap.interpretIng(eachAn)
            if (retVal):
                instrctXWithOri.append((allTexts[i], retVal))

        elif mngTgDat.isComb(eachAn):
            if verbose:
                print "comb", eachAn
            retVal = ingredientMap.interpretComb(eachAn)
            if (retVal):
                instrctXWithOri.append((allTexts[i], retVal))


            global countCombTags
            countCombTags += 1


        elif mngTgDat.isMix(eachAn):
            if verbose:
                print "mix", eachAn
            retVal = ingredientMap.interpretMix(eachAn)
            if (retVal):
                instrctXWithOri.append((allTexts[i], retVal))
                
        elif mngTgDat.isCut(eachAn):
            if verbose:
                print "cut", eachAn
            retVal = ingredientMap.interpretCut(eachAn)
            if (retVal):
                instrctXWithOri.append((allTexts[i], retVal))
                
        elif mngTgDat.isDo(eachAn):
            if verbose:
                print "do", eachAn
            retVal = ingredientMap.interpretDo(eachAn)
            if (retVal):
                instrctXWithOri.append((allTexts[i], retVal))
                
        elif mngTgDat.isCut(eachAn):
            if verbose:
                print "cut", eachAn
            retVal = ingredientMap.interpretCut(eachAn)
            if (retVal):
                instrctXWithOri.append((allTexts[i], retVal))
                
        elif mngTgDat.isCook(eachAn):
            if verbose:
                print "cook", eachAn
            retVal = ingredientMap.interpretCook(eachAn)
            if (retVal):
                instrctXWithOri.append((allTexts[i], retVal))
        
        elif mngTgDat.isPut(eachAn):
            if verbose:
                print "put", eachAn
            retVal = ingredientMap.interpretPut(eachAn)
            if (retVal):
                instrctXWithOri.append((allTexts[i], retVal))
                
        elif mngTgDat.isSep(eachAn):
            if verbose:
                print "separate", eachAn
            global countSepTags
            countSepTags += 1
            retVal = ingredientMap.interpretSep(eachAn)
            if (retVal):
                instrctXWithOri.append((allTexts[i], retVal))
        
            
    return instrctXWithOri, ingredientMap

def displayGraph(ingredientMap):
    # Display graph
    # list of foods
    foodNodes = ingredientMap.m.values()
    # table for display
    numNodeComb = 0
    numNodeSep = 0
    displayTable = [ [row,"nameFood",[]] for row in range(len(foodNodes)) ]
    for eachFood in foodNodes:
        displayTable[eachFood.getIndex()][1] = eachFood.getName()
    for i,each in enumerate(ingredientMap.adjList):
        stringChild = [str(eachChild) for eachChild in each ]
        if (len(stringChild) > 1):
            numNodeSep +=1
        elif (len(stringChild) == 1):
            numNodeComb += 1
        stringChild = ",".join(stringChild)
        displayTable[i][2] = stringChild
    global countComb, countSep
    countComb += numNodeComb
    countSep += numNodeSep
        
    print tabulate(displayTable, headers=["node-id","node-form", "child-id"])
    
#     originalTextsWithEdge = []
#     for each in originalTexts:
#         if each in dictTextEdge.keys():
#             originalTextsWithEdge.append("\nEDGE:"+each +'\n'+ str(dictTextEdge[each])+"\n")
#         else:
#             originalTextsWithEdge.append(each)
    
     
    
```

```python
from nltk.parse.stanford import StanfordParser
english_parser = StanfordParser('/Users/Mick/Documents/workspace/stanford-parser-full-2014-10-31/stanford-parser.jar', '/Users/Mick/Documents/workspace/stanford-parser-full-2014-10-31/stanford-parser-3.5.0-models.jar')
def stanfordTag(sentence):
    global english_parser
    tree = english_parser.raw_parse_sents(("",sentence))
    return tree[0]

def getNounsDes(POSTree):

    nouns = []
    for stree in POSTree.subtrees():
        if (stree.label() == "NP" and stree.height() == 3):
            pos = stree.pos()
            numTags = len(pos)
            j = 0
            while (j < numTags):
                newWord = []
                if (pos[j][1][:2] == "NN" or (pos[j][1][:2] == "JJ" and j != numTags-1)):
                    newWord.append(pos[j][0])
                    j += 1
                    if (j >= numTags):
                        nouns.append(" ".join(newWord))
                        break
                    while (pos[j][1][:2] == "NN"):
                        newWord.append(pos[j][0])
                        j += 1
                        if (j >= numTags):
                            break
                    nouns.append(" ".join(newWord))
                else:
                    j += 1
    return nouns
        
    
```

```python
def buildMilkBaseline(ingloca_List_List, remainingNoun_List_List, ingGlobal_List):
    
    # ingloca_List_List = [[(name, index),(name,index)],[],...]
    # remainingNoun_List_List = [[name, name],[],...]
    # ingGlobal_list = [name, name, name]
    assert len(ingloca_List_List) == len(remainingNoun_List_List)
    
    adjListIng = [[] for i in range(len(ingGlobal_List))]
    availableIng = [None for i in range(len(ingGlobal_List))]
    intermediate = [] # keep track of intermediate ings made
    availableIntermediate = [] # keep track of availability of ings made
    
    #!! currently does not try to link up the intermediate eventhough some have 
    ## this refers to the else: part when the ingredients are called for the second time 
    ## - > then we would check the list of intermediate food made
    
    for i, ingloc_list in enumerate(ingloca_List_List):
        if len(ingloc_list) <= 1:
            continue
        outputIng = remainingNoun_List_List[i]
        inputIng = []
        # check if input refers to anything from the immediate ing
        
        newingloc_list = []
        checkRepeat = {}
        for each_ing in ingloc_list:
            if (each_ing[1] not in checkRepeat.keys()):                
                newingloc_list.append(each_ing)
                checkRepeat[each_ing[1]] = 0
        ingloc_list = newingloc_list
                
        
        # mark the ingredients to be unvail
        for j, each_ing in enumerate(ingloc_list):
            # if it's available the list - bring to new node
            (foodName, idxToList) = each_ing

            if (availableIng[idxToList] == None):
                availableIng[idxToList] = len(intermediate)+ len(ingGlobal_List)
                inputIng.append(each_ing)
            else:

                # (referred afterwards if it's in the intermediate)
                # actual reference to the intermeidate should be substracted with length of given ing
#                 print "intermed: ", intermediate
#                 print "indexToList: ", idxToList
#                 print "available: ", availableIng
#                 print "resl: ", availableIng[idxToList]-len(ingGlobal_List)
                
                inputIng.append((",".join([d[0] for d in intermediate[availableIng[idxToList]-len(ingGlobal_List)]]), availableIng[idxToList] ))
            
        intermediate.append(inputIng)
#             outputIng.append(foodName)
    
    print "intermediate: ", intermediate
    print "availableIng: ", availableIng

    displayTable = [ [row,"nameFood",[]] for row in range(len(availableIng)+len(intermediate)) ]
    for i, childNode in enumerate(availableIng):
        displayTable[i][2] = childNode
        displayTable[i][1]  = '"'+ingGlobal_List[i]+'"'
    for i in range(len(availableIng),len(availableIng)+len(intermediate)):
        displayTable[i][1]  = '"'+"+".join([d[0] for d in intermediate[i-len(availableIng)]])+'"'
        displayTable[i][2]  = None

    print tabulate(displayTable, headers=["node-id","node-form", "child-id"])
    
    return len(intermediate)
            
     
```

```python
def retrieveIngPOStoInglist(myStem, sentence,ingList):
    
    tree = stanfordTag(eachInstruct.lower().rstrip().rstrip("."))
    potentialIng = getNounsDes(tree)
    allIngIndexFromInstructions = []
    allIngFromInstructions = []
    if potentialIng:
        for eachPotentialIng in potentialIng:
            finalRes = prsIngHp.checkNGramMatch(myStem, eachPotentialIng,ingList)
            finalRes = list(set((finalRes)))
            allIngIndexFromInstructions += [each[3] for each in finalRes]
            ingSet = prsIngHp.getOriginWordFromSentence(eachPotentialIng, [(each[1],each[2],each[3],each[4]) for each in finalRes])
            ingWithIndex = [(ingSet[j], each[3]) for j, each in enumerate(finalRes)]
            ingWithIndex = list(set(ingWithIndex))
            allIngFromInstructions += ingSet
    for i, each in enumerate(allIngFromInstructions):
        print [allIngIndexFromInstructions[i], each],
    print
    return allIngIndexFromInstructions, allIngFromInstructions


def chooseWords(eachInstruct, finalRes, ingList, checkIngIfUsed):
    
    ingSet = prsIngHp.getOriginWordFromSentence(eachInstruct, [(each[1],each[2],each[3],each[4]) for each in finalRes])
    ingWithIndex = [(ingSet[j], each[3]) for j, each in enumerate(finalRes)]
    
    ## aggregate index of items that activate multiple hits
    countIng = {}
    for each in ingWithIndex:
        wordIng = each[0]
        indexIng = each[1]
        if wordIng in countIng.keys():
            countIng[wordIng].append(indexIng)
        else:
            countIng[wordIng] = [indexIng]
    
    ## rank the multiple hits
    
    for each in countIng.keys():
        if len(countIng[each])>1:
            # rank all the potential ingredients given the hit 
            rankedWords = process.extract(each, [ingList[indexEach] for indexEach in countIng[each]])
            countIng[each] = rankedWords
        else:
            # the only hit - give ranking 101
            countIng[each] = [(ingList[indexEach],101) for indexEach in countIng[each] ]
            
    ## iterate through the original words hit
    setIngSet = set(ingSet)
    retreivedIng = []
    ingListClone = [ingList[j] if each else "" for j, each in enumerate(checkIngIfUsed)]
    for originHit in countIng.keys():
#         print "originHit", originHit, countIng[originHit]
        for eachHit in countIng[originHit]:
            if eachHit[0] in ingListClone:
                indexOfBestHit = ingListClone.index(eachHit[0])
                if checkIngIfUsed[indexOfBestHit]:
                    checkIngIfUsed[indexOfBestHit] = False
                    ingListClone[indexOfBestHit] = ""
                    retreivedIng.append(indexOfBestHit)
                    break

        
#     print "remaining", [ingList[i] for i, each in enumerate(checkIngIfUsed) if each]
#     print [ingList[each] for each in retreivedIng]
    
    return retreivedIng
```
