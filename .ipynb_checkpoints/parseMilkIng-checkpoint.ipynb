```python
def findRepeatedWord(Y):
    
    allYs = ""
    for y in Y:
        print y
        allYs += y
    words = word_tokenize(allYs)
    fdist = FreqDist(words)
    return fdist

def getSyntaxTree(sentence):
    tokens = word_tokenize(sentence)
    tagged = pos_tag(tokens)
    entities = chunk.ne_chunk(tagged)
    print entities
    return entities

def removeThesePOS(sentence, POStoRemoved):

    tokens = word_tokenize(sentence)
    tagged = pos_tag(tokens)
    tagged = removeVBD(tagged)
    remainTok = [d[0] for d in tagged if not d[1] in POStoRemoved]
    return " ".join(remainTok)

def removeVBD(taggedToks):
    return [d for d in taggedToks if not (d[1] == 'VBD' and len(d[0])>=5 and d[0][-2]+d[0][-1] == "ed")]    

def normalizeIng(ingString):
            
        hasAny = 0
        status = ''
        ingList = []

        # bigram matching
        textIng = word_tokenize(ingString)
        unstemmedTextIng = textIng
        textIng = [myStem.stem(eachTok) for eachTok in textIng ]
        numIng = [i for i in range(len(textIng))  if (textIng[i] not in units and not textIng[i].isdigit())]

        unstemmedTextIng = [ unstemmedTextIng[eachI] for eachI in numIng ]

        textIng = [each for each in textIng if (each not in units and not each.isdigit())]
        bigrams = ngrams(textIng, 2)
        for eachBg in bigrams:
            bgString = " ".join([eachBg[i] for i in range(len(eachBg))])
            if bgString in ingDict:
                hasAny = 1
                status = "matchBi"
                ingList.append(bgString)

        if (hasAny == 0):
            for each in textIng:
                if each in ingDict:
                    hasAny = 1
                    status = "matchIndiv"
                    ingList.append(each)

        if (hasAny == 1):
            ingDesc = ' '.join(ingList)
            ingDesc = '\"'+ingDesc+'\"'
        else:
            status = "matchNone"
            ingDesc = ' '.join(unstemmedTextIng)
    #         dbConcept = dbSpotLightAnnotate(ingDesc,None,None,None)

            global nomatchIngAll
    #         nomatchIngAll.append((ingDesc,dbConcept))
            nomatchIngAll.append(ingDesc)
    #         print (ingDesc)

        return ingDesc, status
```

```python
def dbSpotLightAnnotate(text,conf,support,types):
#     print text
#     text = "package frozen mixed berries, sliced peaches, drained 2 tablespoons honey. In a blender, combine frozen fruit, canned fruit and honey. Blend until smooth. Add chicken and pork." 
    conf = 0.1
    support = 20
    types = "Freebase:/food/food,Freebase:/food/ingredient"
    parameters = {}
    parameters["text"] = text
    parameters["confidence"] = conf
    parameters["support"] = support
    parameters["types"] = types
    headers = {"Accept":'application/json'}
    r = requests.get('http://spotlight.dbpedia.org/rest/annotate', params = parameters, headers=headers)
    queriedRes = r.json()
#     print "got data"
#     print queriedRes
    concept_pos = []
    for eachTag in queriedRes[u'Resources']:
        originalWord = eachTag[u'@surfaceForm']
        dbConcept = eachTag[u'@URI'].split("/")[-1]
        pos = eachTag[u'@offset']
        concept_pos.append((originalWord,dbConcept,pos))
#         print originalWord, dbConcept,pos
#         print eachTag[u'@URI']
    return concept_pos
    
```

# Stanford tagging

```python
from nltk.parse.stanford import StanfordParser
english_parser = StanfordParser('/Users/Mick/Documents/workspace/stanford-parser-full-2014-10-31/stanford-parser.jar', '/Users/Mick/Documents/workspace/stanford-parser-full-2014-10-31/stanford-parser-3.5.0-models.jar')
def stanfordTag(sentence):
    global english_parser
    tree = english_parser.raw_parse_sents(("",sentence))
    return tree[0]

def getNounsDes(POSTree):

    nouns = []
    for stree in POSTree.subtrees():
        if (stree.label() == "NP" and stree.height() == 3):
            pos = stree.pos()
            numTags = len(pos)
            j = 0
            while (j < numTags):
                newWord = []
                if (pos[j][1][:2] == "NN" or (pos[j][1][:2] == "JJ" and j != numTags-1)):
                    newWord.append(pos[j][0])
                    j += 1
                    if (j >= numTags):
                        nouns.append(" ".join(newWord))
                        break
                    while (pos[j][1][:2] == "NN"):
                        newWord.append(pos[j][0])
                        j += 1
                        if (j >= numTags):
                            break
                    nouns.append(" ".join(newWord))
                else:
                    j += 1
    return nouns
        
    
```

```python
def buildMilkBaseline(ingloca_List_List, remainingNoun_List_List, ingGlobal_List):
    
    # ingloca_List_List = [[(name, index),(name,index)],[],...]
    # remainingNoun_List_List = [[name, name],[],...]
    # ingGlobal_list = [name, name, name]
    assert len(ingloca_List_List) == len(remainingNoun_List_List)
    
    adjListIng = [[] for i in range(len(ingGlobal_List))]
    availableIng = [None for i in range(len(ingGlobal_List))]
    intermediate = [] # keep track of intermediate ings made
    availableIntermediate = [] # keep track of availability of ings made
    
    #!! currently does not try to link up the intermediate eventhough some have 
    ## this refers to the else: part when the ingredients are called for the second time 
    ## - > then we would check the list of intermediate food made
    
    for i, ingloc_list in enumerate(ingloca_List_List):
        if len(ingloc_list) <= 1:
            continue
        outputIng = remainingNoun_List_List[i]
        inputIng = []
        # check if input refers to anything from the immediate ing
        
        newingloc_list = []
        checkRepeat = {}
        for each_ing in ingloc_list:
            if (each_ing[1] not in checkRepeat.keys()):                
                newingloc_list.append(each_ing)
                checkRepeat[each_ing[1]] = 0
        ingloc_list = newingloc_list
                
        
        # mark the ingredients to be unvail
        for j, each_ing in enumerate(ingloc_list):
            # if it's available the list - bring to new node
            (foodName, idxToList) = each_ing

            if (availableIng[idxToList] == None):
                availableIng[idxToList] = len(intermediate)+ len(ingGlobal_List)
                inputIng.append(each_ing)
            else:

                # (referred afterwards if it's in the intermediate)
                # actual reference to the intermeidate should be substracted with length of given ing
#                 print "intermed: ", intermediate
#                 print "indexToList: ", idxToList
#                 print "available: ", availableIng
#                 print "resl: ", availableIng[idxToList]-len(ingGlobal_List)
                
                inputIng.append((",".join([d[0] for d in intermediate[availableIng[idxToList]-len(ingGlobal_List)]]), availableIng[idxToList] ))
            
        intermediate.append(inputIng)
#             outputIng.append(foodName)
    
    print "intermediate: ", intermediate
    print "availableIng: ", availableIng

    displayTable = [ [row,"nameFood",[]] for row in range(len(availableIng)+len(intermediate)) ]
    for i, childNode in enumerate(availableIng):
        displayTable[i][2] = childNode
        displayTable[i][1]  = '"'+ingGlobal_List[i]+'"'
    for i in range(len(availableIng),len(availableIng)+len(intermediate)):
        displayTable[i][1]  = '"'+"+".join([d[0] for d in intermediate[i-len(availableIng)]])+'"'
        displayTable[i][2]  = None

    print tabulate(displayTable, headers=["node-id","node-form", "child-id"])
    
    return len(intermediate)
            
     
```

```python
def retrieveIngPOStoInglist(myStem, sentence,ingList):
    
    tree = stanfordTag(eachInstruct.lower().rstrip().rstrip("."))
    potentialIng = getNounsDes(tree)
    allIngIndexFromInstructions = []
    allIngFromInstructions = []
    if potentialIng:
        for eachPotentialIng in potentialIng:
            finalRes = prsIngHp.checkNGramMatch(myStem, eachPotentialIng,ingList)
            finalRes = list(set((finalRes)))
            allIngIndexFromInstructions += [each[3] for each in finalRes]
            ingSet = prsIngHp.getOriginWordFromSentence(eachPotentialIng, [(each[1],each[2],each[3],each[4]) for each in finalRes])
            ingWithIndex = [(ingSet[j], each[3]) for j, each in enumerate(finalRes)]
            ingWithIndex = list(set(ingWithIndex))
            allIngFromInstructions += ingSet
    for i, each in enumerate(allIngFromInstructions):
        print [allIngIndexFromInstructions[i], each],
    print
    return allIngIndexFromInstructions, allIngFromInstructions


def chooseWords(eachInstruct, finalRes, ingList, checkIngIfUsed):
    
    ingSet = prsIngHp.getOriginWordFromSentence(eachInstruct, [(each[1],each[2],each[3],each[4]) for each in finalRes])
    ingWithIndex = [(ingSet[j], each[3]) for j, each in enumerate(finalRes)]
    
    ## aggregate index of items that activate multiple hits
    countIng = {}
    for each in ingWithIndex:
        wordIng = each[0]
        indexIng = each[1]
        if wordIng in countIng.keys():
            countIng[wordIng].append(indexIng)
        else:
            countIng[wordIng] = [indexIng]
    
    ## rank the multiple hits
    
    for each in countIng.keys():
        if len(countIng[each])>1:
            # rank all the potential ingredients given the hit 
            rankedWords = process.extract(each, [ingList[indexEach] for indexEach in countIng[each]])
            countIng[each] = rankedWords
        else:
            # the only hit - give ranking 101
            countIng[each] = [(ingList[indexEach],101) for indexEach in countIng[each] ]
            
    ## iterate through the original words hit
    setIngSet = set(ingSet)
    retreivedIng = []
    ingListClone = [ingList[j] if each else "" for j, each in enumerate(checkIngIfUsed)]
    for originHit in countIng.keys():
#         print "originHit", originHit, countIng[originHit]
        for eachHit in countIng[originHit]:
            if eachHit[0] in ingListClone:
                indexOfBestHit = ingListClone.index(eachHit[0])
                if checkIngIfUsed[indexOfBestHit]:
                    checkIngIfUsed[indexOfBestHit] = False
                    ingListClone[indexOfBestHit] = ""
                    retreivedIng.append(indexOfBestHit)
                    break

        
#     print "remaining", [ingList[i] for i, each in enumerate(checkIngIfUsed) if each]
#     print [ingList[each] for each in retreivedIng]
    
    return retreivedIng
```
