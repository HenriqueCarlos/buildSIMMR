```python
def groupThisSize(groupSize, dataToBeGrouped):
    assert len(dataToBeGrouped) == sum(groupSize)
    posDat = 0
    groupedDat = []
    for size in groupSize:
        thisGroup = []
        for i in range(size):
            thisGroup.append(dataToBeGrouped[posDat])
            posDat += 1
        groupedDat.append(thisGroup)
    return groupedDat

def degroupFindMax(groupedData):
    degrouped = []
    for group in groupedData:
        posMax = group.index(max(group))
        groupWithMax = [0 for i in range(len(group))]
        groupWithMax[posMax] = 1
        degrouped +=  groupWithMax
    return degrouped

def evalRes(pred, ytarget):
    
    posCrrect = 0
    negCrrect = 0
    posWrng = 0
    negWrng = 0
    for i, each in enumerate(pred):
        if pred[i] >= 0.5 and ytarget[i] == 1:
            posCrrect += 1
        elif pred[i] < 0.5 and ytarget[i] == 0:
            negCrrect += 1
        elif pred[i] >= 0.5 and ytarget[i] == 0:
            posWrng += 1
        elif pred[i] < 0.5 and ytarget[i] == 1:
            negWrng += 1
    acc = (posCrrect+negCrrect)/float(len(allYVal))
    truePos = posCrrect/float(posWrng+posCrrect)
    trueNeg = negCrrect/float(negWrng+negCrrect)
    print "total acc: ", acc
    print "true pos: ", truePos
    print "true neg: ", trueNeg
    
    return acc, truePos, trueNeg
    
    
```

```python
testSVMRank = 1
if testSVMRank == 1:
    processDat = 1
    evalDat = 1
    evalModel = 1

def svmRowStringify(dat):
    #1:1 2:1 3:0 4:0.2 5:0
    row = ""
    for i,each in enumerate(dat):
        if each != 0:
            row += " "+str(i+1)+":"+str(each)
    return row

def writeSVMRankFormat(fileName,instructSize,allX,allY):
    SVMRankFile = open(fileName, "w")
    ingQueries = groupThisSize(instructSize, allX)
    ingResult = groupThisSize(instructSize, allY)
    for i,ing in enumerate(ingQueries):
        thisRank = ""
        for j, instrc in enumerate(ing):
            thisDat = ""
            thisDat += str(ingResult[i][j]+1)
            thisDat += " qid:"+str(i+1)
            thisDat += svmRowStringify(instrc)
            thisDat += " #"
            thisRank += thisDat+"\n"
        SVMRankFile.write(thisRank)
    SVMRankFile.close()
```
